```python
prefix = 'NO'
path=['circular','Scircular','elliptic','Selliptic','solar_earth','Ssolar_earth','Kozai','SKozai']

data =[np.loadtxt(prefix+name+'.dat') for name in path]

eng =[np.loadtxt(prefix+name+'.err') for name in path]

performance =[np.loadtxt(prefix+name+'.errcurve') for name in path]

def drawErr(errfile):

    fig=plt.figure(figsize=(2*len(errfile),8))
    for i in range(len(errfile)):
        ax = fig.add_subplot(len(errfile)/2,4,2*i+1)
        t = errfile[i][:,0]
        err =  errfile[i][:,1]
        rej = errfile[i][:,2]
        ax.semilogy(t,np.abs(err),'r.',markersize=0.5,linewidth=0.1)
        ax = fig.add_subplot(len(errfile)/2,4,2*i+2)
        ax.plot(t,rej)
    plt.tight_layout()
    plt.savefig(prefix+'test.png',bbox_inches='tight',dpi=300)

def drawPerformance(errfile):

    fig=plt.figure(figsize=(2*len(errfile),8))
    for i in range(len(errfile)):
        ax = fig.add_subplot(len(errfile)/2,4,2*i+1)
        tol = errfile[i][:,0]
        err =  errfile[i][:,1]
        time = errfile[i][:,2]
        ax.loglog(tol,np.abs(err),'r.-')
        ax = fig.add_subplot(len(errfile)/2,4,2*i+2)
        ax.loglog(tol,time)
    plt.tight_layout()
    plt.savefig(prefix+'performance.png',bbox_inches='tight',dpi=300)
    
drawErr(eng)
drawPerformance(performance)
```
